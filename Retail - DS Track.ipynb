```python
cols_to_convert = ['Invoice', 'StockCode', 'Quantity', 'Price', 'Customer ID']

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['InvoiceDate']= pd.to_datetime(df['InvoiceDate'])
```

```python
import pandas as pd
from datetime import datetime

# Calculate the Monetary value for each transaction
df['TotalAmount'] = df['Quantity'] * df['Price']

# Set the latest date for calculating recency
latest_date = datetime.now()

# Group by Customer ID and calculate RFM values
rfm = df.groupby('Customer ID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,  # Recency: days since last purchase
    'Invoice': 'count',                                     # Frequency: number of purchases
    'TotalAmount': 'sum'                                    # Monetary: total amount spent
})

# Rename columns
rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'Invoice': 'Frequency',
    'TotalAmount': 'Monetary'
}, inplace=True)

print(rfm.head())
```

```python
rfm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 3, figsize=(16, 8))

# Recency distribution plot
sns.histplot(rfm['Recency'], kde=True, ax=ax[0])

# Frequency distribution plot
sns.histplot(rfm.query('Frequency < 1000')['Frequency'], kde=True, ax=ax[1])

# Monetary distribution plot
sns.histplot(rfm.query('Monetary < 10000')['Monetary'], kde=True, ax=ax[2])
```

```python
# Identify the three segments (33% in each segment)
segments = [0.33, 0.66]

quantiles = rfm[["Recency", "Frequency", "Monetary"]].quantile(q=segments).values


# Assign individual R, F, M scores to the three segments of customers

# R Score
rfm['R_Score'] = np.where(rfm['Recency'] < quantiles[0][0], 1, 
                np.where(rfm['Recency'] < quantiles[1][0], 2, 3))

# F Score
rfm['F_Score'] = np.where(rfm['Frequency'] < quantiles[0][1], 3, 
                np.where(rfm['Frequency'] < quantiles[1][1], 2, 1))

# M Score
rfm['M_Score'] = np.where(rfm['Monetary'] < quantiles[0][2], 3, 
                np.where(rfm['Monetary'] < quantiles[1][2], 2, 1))

# Calculate the combined RFM Score
rfm['RFM_Score'] = rfm['R_Score'] + rfm['F_Score'] + rfm['M_Score']

# Create RFM Score Group
rfm['RFM_ScoreGroup'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
```

```python
# Identify the three segments again based on RFMScore (33% in each segment)
segments = [0.33, 0.66]
quantiles = rfm["RFM_Score"].quantile(segments).values

# Define loyalty levels
loyalty_level = ['Platinum', 'Gold', 'Silver']

# Assign loyalty level based on RFM_Score
rfm_loyalty = rfm.assign(
    Loyalty=np.where(rfm["RFM_Score"] <= quantiles[0], loyalty_level[0],
             np.where(rfm["RFM_Score"] <= quantiles[1], loyalty_level[1], 
             loyalty_level[2]))
)
```
